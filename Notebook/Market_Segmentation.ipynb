# Scaling the dataFrame

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled_df = sc.fit_transform(df)
```

```python
scaled_df
```

# Dimensionality reduction
Converting the DataFrame into 2D DataFrame for a better visualization

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

# Hyperparameter tuning
## Finding 'k' value by Elbow Method

```python
from sklearn.cluster import KMeans
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

If we observe the elbow plot there is a slight turn at 4 that's why taking the k value as 4 

# Model Building using KMEANS

```python
kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

```python
labels = kmeans_model.labels_
```

```python
labels
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
visualizer = SilhouetteVisualizer(kmeans_model, colors='yellowbrick')
visualizer.fit(scaled_df)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

# Visualizing the clustered dataframe

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue','yellow'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

# Model Building based on Silhoutte score

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
range_val = range(2,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    
    silhouette_scores.append(silhouette_score(scaled_df, kmean.fit_predict(scaled_df)))
plt.plot(range_val,silhouette_scores,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sihoutte Scores') 
plt.title('Deciding clusters based on Silhoutte score') 
plt.show()
```

if we observe the above graph we can see that the silhoutte score is better hen k = 3

# Building and Visualizing the different clustering models for different values of k

```python
ac2 = AgglomerativeClustering(n_clusters = 2)

# Visualizing the clustering
plt.figure(figsize =(6, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'],
c = ac2.fit_predict(scaled_df), cmap ='rainbow')
plt.show()
```

```python
ac3 = AgglomerativeClustering(n_clusters = 3)

# Visualizing the clustering
plt.figure(figsize =(6, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'],
c = ac3.fit_predict(scaled_df), cmap ='rainbow')
plt.show()
```

```python
ac4 = AgglomerativeClustering(n_clusters = 4)

# Visualizing the clustering
plt.figure(figsize =(6, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'],
c = ac4.fit_predict(scaled_df), cmap ='rainbow')
plt.show()
```

```python
ac5 = AgglomerativeClustering(n_clusters = 5)

# Visualizing the clustering
plt.figure(figsize =(6, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'],
c = ac5.fit_predict(scaled_df), cmap ='rainbow')
plt.show()
```

```python
ac6 = AgglomerativeClustering(n_clusters = 6)

# Visualizing the clustering
plt.figure(figsize =(6, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'],
c = ac6.fit_predict(scaled_df), cmap ='rainbow')
plt.show()
```

```python
k = [2, 3, 4, 5, 6]

# Appending the silhouette scores of the different models to the list
silhouette_scores = []
silhouette_scores.append(
        silhouette_score(scaled_df, ac2.fit_predict(scaled_df)))
silhouette_scores.append(
        silhouette_score(scaled_df, ac3.fit_predict(scaled_df)))
silhouette_scores.append(
        silhouette_score(scaled_df, ac4.fit_predict(scaled_df)))
silhouette_scores.append(
        silhouette_score(scaled_df, ac5.fit_predict(scaled_df)))
silhouette_scores.append(
        silhouette_score(scaled_df, ac6.fit_predict(scaled_df)))

# Plotting a bar graph to compare the results
plt.bar(k, silhouette_scores)
plt.xlabel('Number of clusters', fontsize = 20)
plt.ylabel('S(i)', fontsize = 20)
plt.show()
```

If we observe the above bar chart even the Agglomerative clustering was not able to divide the groups precisely
